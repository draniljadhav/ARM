```python
#import all required classes for market basket analysis from mlxtedn library
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Example 2

In this example instead of reading data from external sources we are going to 
create small transaction dataset to understand how to generate rules from
sample transaction dataset using apriori algorithm.

This example is based on data and document by Sebastian Raschka

```python
transaction_data = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
#display transactio data
transaction_data
```

Now transform data into dataframe. Each row in dataframe represents transaction and each columms represent whether item present in transaction.

```python
#create TransactionEncoder object. This is required to bring data in the required format  
transaction_encoder = TransactionEncoder()
```

```python
#fit and transform data
transaction_array = transaction_encoder.fit(transaction_data).transform(transaction_data)
```

```python
#display array
transaction_array
```

```python
#check columns
transaction_encoder.columns_
```

```python
#import pandas library
import pandas as pd
```

```python
#Create dataframe
transaction_data = pd.DataFrame(transaction_array, columns=transaction_encoder.columns_)
```

```python
#display data
transaction_data
```

```python
#find dimension of  data
transaction_data.shape
```

```python
#Generate frequent itemsets with support value 0.6. ie frequent itemset should be part of minimum 60% transactions
frequent_itemsets = apriori(transaction_data, min_support=0.6, use_colnames=True)
#by default it consider column index. use_colnames=True consider column names
```

```python
#display frequent itemsets. 
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)
```

```python
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

Pandas DataFrames make it easy to filter the results further. Let's say we are ony interested in rules 
that satisfy the following criteria: 
1. at least 2 antecedents
2. confidence > 0.75
3. lift score > 1.2

```python
#for each rule find length of antecedents and store it in seperate column in dataframe
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
#apply filter condition
rules[ (rules['antecedent_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```
