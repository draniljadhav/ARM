```python
#import all required classes for market basket analysis from mlxtedn library
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
#import pandas library 
import pandas as pd
```

# Example 5

In this example we will read data from csv file

```python
#read data from file
consumer_data=pd.read_csv("D:/mdp iitb/DataForArules.csv")
```

```python
consumer_data.head(5)
```

```python
#select only required columns
consumer_data=consumer_data.iloc[:,4:13]
```

```python
consumer_data.head(5)
```

```python
#replace yes with True and no with False
consumer_data = consumer_data.applymap(lambda x: True if x == "yes" else False)
```

```python
consumer_data.head(5)
```

```python
#generate frequent itemsets
frequent_itemsets = apriori(consumer_data, min_support=0.3, use_colnames=True)
```

```python
#generate rules
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
len(rules)
```

```python
#display rules
rules
```

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequents_len"] = rules["consequents"].apply(lambda x: len(x))
rules["total_len"] = rules["antecedent_len"]+rules["consequents_len"]
```

```python
rules[ (rules['antecedent_len'] > 1) &
       (rules['total_len'] >2) &
       (rules['consequents_len'] ==1)     ]
```
